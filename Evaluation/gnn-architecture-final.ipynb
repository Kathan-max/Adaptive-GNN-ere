<h1>Again Clustering of Relations

```python
def performOnlineKNN(data, th):
    # df = pd.read_csv(csvp)
    embed_dict = {}
    for _,row in df.iterrows():
        if row['relation'] not in df:
            feat = row['relEmbed']
            embed_dict[row['relation']] = feat
    
    centroids = []
    centroids_cluster = []
    centroid_counts = []
    for rel, embed in tqdm_notebook(embed_dict.items(), total= len(embed_dict), desc = "Processing Embeddings: "):
        if len(centroids) == 0:
            centroids.append(embed)
            centroids_cluster.append([rel])
            centroid_counts.append(1)
        else:
            update = False
            for i, cent in enumerate(centroids):
                sim = cosine_similarity(cent.reshape(1, -1), embed.reshape(1, -1))[0][0]
                if sim >= th:
                    centroids[i] = (centroids[i] * centroid_counts[i] + embed) / (centroid_counts[i] + 1)
                    centroid_counts[i] += 1
                    if len(centroids_cluster[i]) == 0:
                        centroids_cluster[i] = [rel]
                    else:
                        centroids_cluster[i].append(rel)
                    update = True
                    break
            if not update:
                centroids.append(embed)
                centroids_cluster.append([rel])
                centroid_counts.append(1)
    return centroids, centroids_cluster
                    
```

```python
def getName(data, centroids_cluster):
    df = data
    embed_dict = {}
    for _,row in df.iterrows():
        if row['relation'] not in df:
            feat = row['relEmbed']
            embed_dict[row['relation']] = feat
    rel_Cluster_Name = {}
    for clust in tqdm_notebook(centroids_cluster, total = len(centroids_cluster), desc = "Finding BestName"):
        most_sim = ""
        max_sum = 0
        if len(clust) == 1:
            rel_Cluster_Name[clust[0]] = clust[0]
            continue
        for ele in clust:
            sim_sum = 0
            for ele2 in clust:
                if ele == ele2: 
                    continue
                sim = cosine_similarity(embed_dict[ele].reshape(1, -1), embed_dict[ele2].reshape(1, -1))[0][0]
                sim_sum += sim
            if  max_sum < sim_sum:
                max_sum = sim_sum
                most_sim = ele
        for ele in clust:
            rel_Cluster_Name[ele] = most_sim
             
    return rel_Cluster_Name, embed_dict
```

<h2>Building Graph out of CSV

```python
def build_graph_from_csv(df):
    node_dict = {}
    node_features = []
    node_labels = []
    total_uniqueLabels = []
    for lab in df['headType'].unique():
        total_uniqueLabels.append(lab)
    for lab in df['tailType'].unique():
        if lab not in total_uniqueLabels:
            total_uniqueLabels.append(lab)

    org_index = {i:lab_ for i,lab_ in enumerate(total_uniqueLabels)}
    index_org = {lab_:i for i,lab_ in org_index.items()}
    # print(total_uniqueLabels)
    # print(index_org)
    current_idx = 0
    for _, row in df.iterrows():
        for ntype in ['head', 'tail']:
            ent = row[ntype]
            if ent not in node_dict:
                m = row[ntype + "Embed"]
                if(type(m) == str):
                    m = load_embeddings(row[ntype + "Embed"])
                feat = m
                typeStr = row[ntype + "Type"]
                label = index_org[typeStr]
                node_dict[ent] = {
                    "index": current_idx,
                    "feature": feat,
                    "label": label
                }
                
                node_features.append(feat)
                node_labels.append(label)
                current_idx += 1
    print("Total Node labels: ", len(node_labels))
    node_features = torch.tensor(np.vstack(node_features), dtype=torch.float)
    node_labels = torch.tensor(node_labels, dtype=torch.long)

    # building views
    view_edge_dict = {}
    view_edge_attr = {}
    for _, row in df.iterrows():
        view = row["relSecond"]
        head_idx = node_dict[row["head"]]["index"]
        tail_idx = node_dict[row["tail"]]["index"]
        rel_emb = row["relSecondEmbed"]
        if view not in view_edge_dict:
            view_edge_dict[view] = []
            view_edge_attr[view] = []
        view_edge_dict[view].append([head_idx, tail_idx])
        view_edge_attr[view].append(rel_emb)
    view_data = {}
    for view in view_edge_dict.keys():
        edge_index = torch.tensor(view_edge_dict[view], dtype=torch.long).T
        edge_attr = torch.tensor(np.vstack(view_edge_attr[view]), dtype=torch.float)
        # print("edge_index shape: ",edge_index.shape)
        # print("edge_attr shape: ",edge_attr.shape)
        view_data[view] = {
            "edge_index": edge_index, 
            "edge_attr": edge_attr
        }
    
    return node_features, node_labels, view_data, node_dict, index_org
```

<h1>Architecture options for the MVGANS

```python
class GCNBlock(nn.Module):
    def __init__(self, in_channels, hidden_channels, num_classes, dropout):
        super(GCNBlock, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, 384)
        self.conv3 = GCNConv(384, num_classes)
        self.dropout = dropout
        
    def forward(self, x, edge_index):
        hidden_x1 = F.relu(self.conv1(x, edge_index))
        hidden_x2 = F.relu(self.conv2(hidden_x1, edge_index))
        x = F.dropout(hidden_x2, p=self.dropout, training=self.training)
        x = self.conv3(x, edge_index)
        return x, hidden_x2
```

```python
class GATBlock(nn.Module):
    def __init__(self, in_channels, hidden_channels, num_classes, heads, dropout):
        conv2_out_channels = int(384/heads)
        super(GATBlock, self).__init__()
        self.conv1 = GATConv(in_channels, hidden_channels, heads=heads, dropout=dropout)
        self.conv2 = GATConv(hidden_channels*heads, conv2_out_channels, heads=heads, dropout=dropout)
        self.conv3 = GATConv(conv2_out_channels * heads, num_classes, heads=1, dropout=dropout)
        self.dropout = dropout
        
    def forward(self, x, edge_index):
        hidden_x1 = F.relu(self.conv1(x, edge_index))
        hidden_x2 = F.relu(self.conv2(hidden_x1, edge_index))
        x = F.dropout(hidden_x2, p=self.dropout, training=self.training)
        x = self.conv3(x, edge_index)
        return x, hidden_x2
```

```python
class GraphSAGEBlock(nn.Module):
    def __init__(self, in_channels, hidden_channels, num_classes, dropout):
        super(GraphSAGEBlock, self).__init__()
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        self.conv2 = SAGEConv(hidden_channels, 384)
        self.conv3 = SAGEConv(384, num_classes)
        self.dropout = dropout

    def forward(self, x, edge_index):
        hidden_x1 = F.relu(self.conv1(x, edge_index))
        hidden_x2 = F.relu(self.conv2(hidden_x1, edge_index))
        x = F.dropout(hidden_x2, p=self.dropout, training=self.training)
        x = self.conv3(x, edge_index)
        return x, hidden_x2
```

```python
class GINBlock(nn.Module):
    def __init__(self, in_channels, hidden_channels, num_classes, dropout):
        super(GINBlock, self).__init__()
        mlp1 = MLP([in_channels, hidden_channels, hidden_channels])
        mlp2 = MLP([hidden_channels, hidden_channels, 384])
        mlp3 = MLP([384, 384, num_classes])
        self.conv1 = GINConv(mlp1)
        self.conv2 = GINConv(mlp2)
        self.conv3 = GINConv(mlp3)
        self.dropout = dropout

    def forward(self, x, edge_index):
        hidden_x1 = F.relu(self.conv1(x, edge_index))
        hidden_x2 = F.relu(self.conv2(hidden_x1, edge_index))
        x = F.dropout(hidden_x2, p=self.dropout, training=self.training)
        x = self.conv3(x, edge_index)
        return x, hidden_x2
```

```python
SEARCH_SPACE = {
    "model_type": ["GCN", "GAT", "GraphSAGE", "GIN"],
    "hidden_channels": [64, 128],
    # "hidden_channels": [384],
    "dropout": [0.2, 0.3],
    "heads": [2, 4]  # Only relevant for GAT
}
```

```python
def sample_architecture():
    """Sample one candidate architecture from the search space."""
    import random
    arch = {}
    arch["model_type"] = random.choice(SEARCH_SPACE["model_type"])
    arch["hidden_channels"] = random.choice(SEARCH_SPACE["hidden_channels"])
    arch["dropout"] = random.choice(SEARCH_SPACE["dropout"])
    if arch["model_type"] == "GAT":
        arch["heads"] = random.choice(SEARCH_SPACE["heads"])
    else:
        arch["heads"] = None
    return arch
```

<h1>Defining A GNN module (per view)

```python
class ViewGNN(nn.Module):
    def __init__(self, in_channels, arch, num_classes):
        super(ViewGNN, self).__init__()
        hidden = arch["hidden_channels"]
        dropout = arch["dropout"]
        model_type = arch["model_type"]
        if model_type == "GCN":
            self.gnn = GCNBlock(in_channels, hidden, num_classes, dropout)
        elif model_type == "GAT":
            heads = arch["heads"]
            self.gnn = GATBlock(in_channels, hidden, num_classes, heads, dropout)
        elif model_type == "GraphSAGE":
            self.gnn = GraphSAGEBlock(in_channels, hidden, num_classes, dropout)
        elif model_type == "GIN":
            self.gnn = GINBlock(in_channels, hidden, num_classes, dropout)
        else:
            raise ValueError("Unsupported model type")
        # Final layer to map to embedding space (for later aggregation)
        # self.fc = nn.Linear(hidden, num_classes)
    
    def forward(self, x, edge_index):
        logits, hidden_x = self.gnn(x, edge_index)
        return logits, hidden_x
```

```python
# def reduce_DimPCA(small_dim, embedding):
#     print("Embedding shape: ", embedding.shape)
#     pca = PCA(n_components = small_dim)
#     embedding_np = embedding.cpu().detach().numpy()
#     reduced_embedding_np = pca.fit_transform(embedding_np)
#     reduced_embedding = torch.from_numpy(reduced_embedding_np).float()
#     return reduced_embedding
    
def compute_loss(logits, node_labels, agg_emb, label_embeddings, alpha):
    # print("Node labels: ",node_labels.shape)
    target_similarity = F.one_hot(node_labels, num_classes=label_embeddings.shape[0]).float()
    ce_loss = F.cross_entropy(logits, target_similarity)
    # print("Logits's shape: ", logits.shape)
    # print("first logit: ", logits[0])
    # 78 x 512
    agg_emb_norm = F.normalize(agg_emb, p=2, dim=-1)
    # print("agg Norm: ", agg_emb_norm.shape)
    # 7 x 384
    label_embeddings_norm = F.normalize(label_embeddings, p=2, dim=-1)
    # print("Label Emb norm: ", label_embeddings_norm.shape)
    # if agg_emb_norm.shape[1] > label_embeddings_norm.shape[1]:
    #     agg_emb_norm = reduce_DimPCA(label_embeddings_norm.shape[1], agg_emb_norm)
    # elif agg_emb_norm.shape[1] < label_embeddings_norm.shape[1]:
    #     label_embeddings_norm = reduce_DimPCA(agg_emb_norm.shape[1], 
    #                                           label_embeddings_norm)
    similarity_matrix = torch.matmul(agg_emb_norm, label_embeddings_norm.T)
    # print("Simi matrix: ", similarity_matrix.shape)
    # print("Target Sim: ", target_similarity[0][0])
    # print("node_label: ", node_labels)
    # print("Target Shape: ",target_similarity.shape)
    zsl_loss = F.mse_loss(similarity_matrix, target_similarity)
    total_loss = (1 - alpha) * ce_loss + alpha * zsl_loss
    return total_loss
```

```python
def train_view_model(model, data, node_features, node_labels, optimizer, label_embeddings, alpha=0.5):
    model.train()
    optimizer.zero_grad()
    # print("Node feature's shape: ", node_features.shape)
    # print("View Data Dict's shape: ",view_data_dict.shape)
    # out = model(node_features, data.edge_index)

    # if not isinstance(view_data, Data):
    #     raise ValueError("view_data must be a PyG Data object")
    # loss = criterion(out, node_labels)
    logits, agg_emb = model(node_features, data.edge_index)
    # print("Agg shape",agg_emb.shape)
    loss = compute_loss(logits, node_labels, agg_emb, label_embeddings, alpha)
    loss.backward()
    optimizer.step()
    return loss.item()

def evaluate_view_model(model, data, node_features, node_labels):
    model.eval()
    with torch.no_grad():
        out, agg_emb = model(node_features, data.edge_index)
        pred = out.argmax(dim=1)
        acc = accuracy_score(node_labels.cpu(), pred.cpu())
        f1 = f1_score(node_labels.cpu(), pred.cpu(), average="macro")
    return acc, f1
```

<h2>Method to search for the best arch for the given view

```python
import random
def search_best_architecture_for_view(view_data, node_features, node_labels, label_embeddings, num_classes, alpha = 0.5, num_trials=5, epochs=20, lr=0.01):
    best_f1 = 0.0
    best_arch = None
    best_model_state = None
    
    
    # edge_index = view_data["edge_index"].long() 
    # edge_attr = view_data["edge_attr"].float()
    # data = Data(x=node_features, edge_index=edge_index, edge_attr=edge_attr)
    data = Data(x=node_features, edge_index=view_data["edge_index"], edge_attr=view_data["edge_attr"])
           
    
    # print("Edge Index shape: ", edge_index.shape)
    # print("Edge Index type: ", edge_index.dtype)
    
    for trial in range(num_trials):
        arch = sample_architecture()
        # Initialize model
        # arch['model_type'] = "GAT"
        # arch["heads"] = random.choice(SEARCH_SPACE["heads"])
        print(f"Trial {trial+1}/{num_trials}, Architecture: {arch}")
        model = ViewGNN(in_channels=node_features.shape[1], arch=arch, num_classes = num_classes)
        model = model.to(device)
        optimizer = optim.Adam(model.parameters(), lr=lr)
        criterion = nn.CrossEntropyLoss()
        
        # Split node indices for train/val
        idx = np.arange(node_features.shape[0])
        train_idx, val_idx = train_test_split(idx, test_size=0.2, random_state=42)
        train_mask = torch.tensor(np.isin(np.arange(len(idx)), train_idx), dtype=torch.bool, device=device)
        val_mask = torch.tensor(np.isin(np.arange(len(idx)), val_idx), dtype=torch.bool, device=device)
        
        for epoch in range(epochs):
            # train_multiview_model(model, view_data_dict, node_features, node_labels, optimizer, label_embeddings, alpha=0.5)
            loss = train_view_model(model, data, node_features, node_labels, optimizer, label_embeddings, alpha = 0.5)
            if (epoch+1) % 5 == 0:
                model.eval()
                with torch.no_grad():
                    out, _ = model(node_features, data.edge_index)
                    pred = out.argmax(dim=1)
                    f1 = f1_score(node_labels[val_mask].cpu(), pred[val_mask].cpu(), average="macro")
                print(f"  Epoch {epoch+1}/{epochs} Loss: {loss:.4f} Val F1: {f1:.4f}")
        
        # Evaluate final model on validation set
        _, f1 = evaluate_view_model(model, data, node_features, node_labels)
        print(f"Trial {trial+1} Final Val F1: {f1:.4f}")
        if f1 > best_f1:
            best_f1 = f1
            best_arch = arch
            best_model_state = model.state_dict()
    
    print("Best Architecture:", best_arch, "with F1:", best_f1)
    # Return best architecture and a fresh model instantiated with that arch and loaded state.
    best_model = ViewGNN(in_channels=node_features.shape[1], arch=best_arch, num_classes = num_classes).to(device)
    best_model.load_state_dict(best_model_state)
    return best_arch, best_model
```

<h1>Multi-View Training

```python
def train_multiview_model(model, view_data_dict, node_features, node_labels, optimizer, criterion):
    model.train()
    optimizer.zero_grad()
    logits = model(node_features, view_data_dict)
    loss = criterion(logits, node_labels)
    loss.backward()
    optimizer.step()
    return loss.item()

def evaluate_multiview_model(model, view_data_dict, node_features, node_labels):
    model.eval()
    with torch.no_grad():
        out = model(node_features, view_data_dict)
        pred = out.argmax(dim=1)
        acc = accuracy_score(node_labels.cpu(), pred.cpu())
        f1 = f1_score(node_labels.cpu(), pred.cpu(), average="macro")
    return acc, f1
```

```python
class MultiViewGNN(nn.Module):
    def __init__(self, view_models, num_views, num_classes):
        super(MultiViewGNN, self).__init__()
        self.view_models = nn.ModuleDict(view_models)
        # # Freezing the parameters of the view models since they are already trained using the few shot and zero shot learning
        # for param in self.view_models.parameters():
        #     param.requires_grad = False
        self.view_weights = nn.Parameter(torch.randn(num_views, num_classes), requires_grad=True)
        self.fc1 = nn.Linear(num_classes, num_classes)
        self.fc2 = nn.Linear(num_classes, num_classes)
    
    def forward(self, x, view_data_dict):
        view_outputs = []
        for view_name, model in self.view_models.items():
            data = view_data_dict[view_name]
            # Assume each view_data is a Data object with edge_index
            logits, _ = model(x, data.edge_index)
            view_outputs.append(logits)

        stacked_views = torch.stack(view_outputs, dim=0)
        stacked_views = stacked_views.permute(1, 0, 2)
        weighted_views = torch.matmul(self.view_weights.T, stacked_views)
        aggregated = weighted_views.sum(dim=1)

        aggregated = F.relu(self.fc1(aggregated))
        aggregated = self.fc2(aggregated)
        return aggregated
```

```python
class MultiViewGNN_Agg(nn.Module):
    def __init__(self, view_models, num_views, num_classes):
        super(MultiViewGNN_Agg, self).__init__()
        self.view_models = nn.ModuleDict(view_models)
        self.fc1 = nn.Linear(num_classes, num_classes)
        self.fc2 = nn.Linear(num_classes, num_classes)
    
    def forward(self, x, view_data_dict):
        view_outputs = []
        for view_name, model in self.view_models.items():
            data = view_data_dict[view_name]
            # Assume each view_data is a Data object with edge_index
            logits, _ = model(x, data.edge_index)
            view_outputs.append(logits)

        # Simple sum of view outputs instead of weighted matrix
        aggregated = torch.stack(view_outputs, dim=0).sum(dim=0)

        aggregated = F.relu(self.fc1(aggregated))
        aggregated = self.fc2(aggregated)
        return aggregated
```
