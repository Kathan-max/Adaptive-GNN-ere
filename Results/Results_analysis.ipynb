```python
def find_uniqmnodes(df):    
    heads = df['head'].unique()
    tails = df['tail'].unique()
    overall = set()
    for head in heads:
        overall.add(head)
    for tail in tails:
        overall.add(tail)
    
    overall1 = set()
    headTypes = df['headType'].unique()
    tailTypes = df['tailType'].unique()
    for headType in headTypes:
        overall1.add(headType)    
    for tailType in tailTypes:
        overall1.add(tailType)
    return len(overall), len(overall1), overall1
```

```python
totalNodes, totalClasses, classes = find_uniqmnodes(fin)
print(totalNodes, totalClasses, classes)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {
    "epochs": [1, 2, 3, 4, 5],
    "f1": [0.7, 0.75, 0.78, 0.8, 0.82],
    "acc": [0.6, 0.68, 0.74, 0.77, 0.79],
    "loss": [0.9, 0.7, 0.5, 0.4, 0.3]
}

df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(df["epochs"], df["f1"], marker='o', label="F1 Score")
plt.plot(df["epochs"], df["acc"], marker='s', label="Accuracy")
plt.plot(df["epochs"], df["loss"], marker='^', label="Loss")

# Labels and title
plt.xlabel("Epochs")
plt.ylabel("Metrics")
plt.title("Model Performance Over Epochs")

# Add legend
plt.legend()

# Show plot
plt.show()
```
